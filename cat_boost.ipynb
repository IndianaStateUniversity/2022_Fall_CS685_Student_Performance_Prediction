```python
from sklearn.compose import ColumnTransformer
from sklearn.datasets import fetch_openml
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

numeric_features=[0,1,2,3,4,5,6,7]
numeric_transformer=Pipeline(steps=[('imputer',IterativeImputer(random_state=1)),('scalar',StandardScaler())])
preprocessor=ColumnTransformer(transformers=[('num',numeric_transformer,numeric_features)])
```

```python
model=Pipeline(steps=[('preprocessor',preprocessor),('classifier',CatBoostClassifier(loss_function='MultiClass',class_weights=[.21,.82,.97,.64,.12,.10,.42,.11,.85,.14,.05,.32,.80,.29,.41,.79,.68,.30],iterations=300,depth=10,random_state=1))]).fit(X_train,y_train)
print(model.score(X_train,y_train))
```
